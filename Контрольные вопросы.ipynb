```python
n = 5 
factor_n = 1 
for i in range(1,n+1):
    factor_n = factor_n * i
print(f"1 Задание : Факториал {n} = {factor_n}")

p = 7
helper = 0
for i in range(1,p+1):
    if (p % i == 0):
        helper +=1
        if (helper>2):
            print(f"2 Задание : Число {p} сложное")
            break
if helper == 2:
    print(f"2 Задание : Число {p} простое")
```
